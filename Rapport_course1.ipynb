# Grille de calcul

## LCG

```python
def LCG(a,b,m,x,n):
    l = [x]
    for i in range(n):
        x=(a*x+b)%m
        l.append(x)
    return l

def park_miller(x,n):
    return LCG(16807,0,pow(2,31) - 1,x,n)
    
```

## Mitchel Moore

```python
def lagged_fibonacci(m, nactual, n, l, j, k):
    while(nactual < n):
        x = (l[nactual - j] + l[nactual - k])%m
        l.append(x)
        nactual = nactual + 1
    return l[55:]

def mitchel_moore(x, n, m):
    l = park_miller(x,55)
    return lagged_fibonacci(m, 55, n + 55, l, 24, 55)
```

## BBS

```python
def BBS(m,x, n):
    l = [x]
    for i in range(n):
        x =x*x%m
        l.append(x)
    return l
```

## Mersene Twister

```python
import random
def twister(n):
    l = []
    for i in range(n):
        l.append(random.random())
    return l
```

```python
import numpy as np
from matplotlib import pyplot as plt 
```

## Mersene Twister - Histogram

```python
plt.rcParams["figure.figsize"] = (20,3)
a = twister(100)
plt.subplot(1,3,1)
plt.hist(a,bins = [0,0.20,0.40,0.60,0.80,1])
plt.title("Histogram Twister 100") 
a = twister(1000)
plt.subplot(1,3,2)
plt.hist(a,bins = [0,0.20,0.40,0.60,0.80,1])
plt.title("Histogram Twister 1000") 
a = twister(100000)
plt.subplot(1,3,3)
plt.hist(a,bins = [0,0.20,0.40,0.60,0.80,1])
plt.title("Histogram Twister 100000") 
plt.show()

print("Q1 quantile of the array : ", np.quantile(a, .25))
print("Q2 quantile of the array : ", np.quantile(a, .50))
print("Q3 quantile of the array : ", np.quantile(a, .75))
print("Variance of the array : ", np.var(a))
```

## Twister - ECDF

```python
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
plt.rcParams["figure.figsize"] = (20,3)

a = mitchel_moore(random.random(), 100,pow(2,31))
e = ECDF(a)
plt.subplot(1,3,1)
plt.plot(e.x, e.y)
plt.title("ECDF twister 100")
a = mitchel_moore(random.random(), 1000,pow(2,31))
e = ECDF(a)
plt.subplot(1,3,2)
plt.plot(e.x, e.y)
plt.title("ECDF twister 1000")
a = mitchel_moore(random.random(), 100000,pow(2,31))
e = ECDF(a)
plt.subplot(1,3,3)
plt.plot(e.x, e.y)
plt.title("ECDF twister 10000")
plt.show()
```
